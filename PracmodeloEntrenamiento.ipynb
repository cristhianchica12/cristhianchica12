```python
url='https://raw.githubusercontent.com/Sinchiguano/ai_googleColab/main/datasetUleamMl/IRIS.csv'
import pandas as pd
dataset=pd.read_csv(url)
```

```python
dataset.head(10)
```

```python
dataset['species'].unique()
```

```python
dataset.shape
```

```python
dataset['species'].value_counts()
```

```python
tem=list()
for etiqueta in dataset['species']:
  if etiqueta =='Iris-setosa':
    tem.append(1)
  elif etiqueta =='Iris-versicolor':
    tem.append(2)
  else:
    tem.append(3)
dataset['etiqueta']=tem
```

```python
dataset.head(10)
```

```python
dataset['etiqueta'].unique()
```

```python
dataset.groupby('species').size()
```

```python
#array(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'], dtype=object)
maskSetosa=dataset['species']=='Iris-setosa'
maskVersicolor=dataset['species']=='Iris-versicolor'
maskVirginica=dataset['species']=='Iris-virginica'


setosa=dataset[maskSetosa]
versicolor=dataset[maskVersicolor]
virginica=dataset[maskVirginica]
```

```python
import matplotlib.pyplot as plt

fig, ax=plt.subplots()
fig.set_size_inches(14,6)


ax.scatter(setosa['sepal_length'],setosa['petal_width'])
ax.scatter(versicolor['sepal_length'],versicolor['petal_width'])
ax.scatter(virginica['sepal_length'],virginica['petal_width'])

ax.legend()
```
